# LangChain Messages

LangChain uses a standardized **message-based interface** to communicate with chat models such as OpenAI, Gemini, Anthropic, Groq, and Ollama. Instead of sending a single string, applications send a list of messages that represent the conversation history. This approach allows the model to understand context, maintain multi-turn conversations, and support advanced capabilities like tool calling, memory, and multimodal interactions. Every chat request is built from one or more message objects, each representing a different role in the conversation.

The most common message types are **SystemMessage**, **HumanMessage**, and **AIMessage**. A `SystemMessage` defines the assistant's behavior, personality, or instructions before the conversation begins. A `HumanMessage` represents input from the user, while an `AIMessage` contains the model's response. During conversations, these messages are stored in sequence, allowing the model to remember previous interactions and generate contextually relevant answers. Most LangChain chat models return an `AIMessage` object that contains not only the generated text but also useful metadata such as token usage, model information, and tool calls.

For applications that interact with external services, LangChain also provides **ToolMessage**, which stores the output of a tool after the model requests one. For example, if an AI agent needs to retrieve the current weather, search a database, or execute code, the tool's result is returned as a `ToolMessage`. The model then reads this message and generates a final response for the user. This structured workflow enables AI agents to combine reasoning with real-world data while keeping the conversation organized and easy to manage.

Messages in LangChain are flexible enough to support multimodal inputs, including text, images, and documents. They also work seamlessly with streaming responses, where partial outputs are received as `AIMessageChunk` objects before being combined into a complete `AIMessage`. Because every interaction is represented as a message, developers can easily implement conversation memory, debugging, logging, and agent workflows. This unified message system is one of the core design principles that makes LangChain a powerful framework for building modern AI applications.

# Text Prompts in LangChain

A text prompt is a set of instructions or input provided to a language model to guide its response. In LangChain, prompts are used to define the task, provide context, and specify the expected output. A prompt can be as simple as a single question or as detailed as a structured instruction with examples, constraints, and formatting requirements. Well-crafted prompts help the model generate more accurate, relevant, and consistent responses.

LangChain provides prompt templates that allow developers to create reusable and dynamic prompts by inserting variables at runtime. This makes it easy to build applications where the prompt changes based on user input, retrieved documents, or conversation history. By combining effective prompt design with LangChain's templating system, developers can create AI applications that are more reliable, maintainable, and capable of handling a wide variety of real-world tasks.

````python
model.invoke("what is langchain?")
````

# When to Use Text Prompts

Use text prompts whenever you need to send a single instruction or task to a language model without maintaining a conversation. They are simple, flexible, and suitable for many AI-powered applications.

Common use cases include:
- Content generation (blogs, emails, articles)
- Text summarization
- Language translation
- Code generation and explanation
- Question answering
- Text classification
- Grammar correction and rewriting
- Data extraction from text
- Sentiment analysis
- Generating structured outputs such as JSON or Markdown

For applications that require multiple turns of conversation or context from previous interactions, chat messages are generally a better choice than a single text prompt.
# Types of Chat Messages

LangChain organizes conversations using different message types, where each message has a specific role. This structured approach helps the model understand who is speaking and how it should respond.

- **SystemMessage** – Defines the assistant's behavior, rules, personality, or instructions before the conversation begins.
- **HumanMessage** – Represents the user's input, such as a question, request, or command.
- **AIMessage** – Contains the language model's response to the user's message.
- **ToolMessage** – Stores the output returned by an external tool after the AI requests it, such as a weather API, database query, or calculator.
- **FunctionMessage** – A legacy message type used for OpenAI function calling. It has largely been replaced by `ToolMessage` in modern LangChain applications.
- **RemoveMessage** – Used in LangGraph to remove messages from conversation history, helping manage memory and control context size.

Together, these message types create a structured conversation history that enables chat models to maintain context, interact with tools, and generate more accurate and coherent responses.

# Message Structure in LangChain

Every message in LangChain is represented as an object containing several fields. While the exact fields may vary depending on the message type and model provider, the following are the most commonly used properties.

| Property | Description |
|----------|-------------|
| **role** | Identifies who sent the message, such as `system`, `user` (human), `assistant` (AI), or `tool`. |
| **content** | The actual text or multimodal content (text, images, documents, etc.) contained in the message. |
| **id** | A unique identifier for the message, useful for tracking, updating, or removing messages. |
| **name** | An optional name for the sender, often used when multiple tools or assistants are involved. |
| **tool_calls** | A list of tool calls requested by the AI, including the tool name and input arguments. Present mainly in `AIMessage`. |
| **invalid_tool_calls** | Stores tool calls that could not be parsed or executed correctly. |
| **response_metadata** | Contains metadata about the model response, such as the model name, finish reason, safety ratings, or provider-specific details. |
| **usage_metadata** | Provides token usage statistics, including input tokens, output tokens, and total tokens. |
| **additional_kwargs** | Stores provider-specific data that is not part of the standard message schema. |

## Example

```python
from langchain_core.messages import HumanMessage

message = HumanMessage(
    content="Explain what LangChain is."
)
```

The model returns an `AIMessage` similar to:

```python
AIMessage(
    content="LangChain is a framework for building AI applications...",
    response_metadata={
        "model_name": "gemini-2.5-flash"
    },
    usage_metadata={
        "input_tokens": 12,
        "output_tokens": 35,
        "total_tokens": 47
    }
)
```

The `role` tells the model **who is speaking**, the `content` contains **what was said**, and the remaining fields provide **metadata, tool information, and usage details** that help developers build reliable and feature-rich AI applications.

````python
from langchain.messages import SystemMessage,HumanMessage,AIMessage
ai_message = AIMessage(content="I can help you with that!", name="Assistant", id="assistant_123456")
messages = [SystemMessage(content="You are a helpful assistant."), human_message, ai_message,HumanMessage(content="What is a Rest API?", name="User", id="user_454578")]
response = model.invoke(messages)
print(response.content)
````

````python
from langchain.messages import SystemMessage,HumanMessage,AIMessage,ToolMessage
ai_message = AIMessage(content=[],tool_calls=[{"name": "get_weather", "args": {"location": "New Delhi"}, "id": "call_123456"}])
weather_result = 'Sunny and 32°C'
tool_message = ToolMessage(content=weather_result, tool_call_id="call_123456")
messages = [SystemMessage(content="You are a helpful assistant."), ai_message, tool_message]
response = model.invoke(messages)
print(response.content)
print(response.usage_metadata)
````
