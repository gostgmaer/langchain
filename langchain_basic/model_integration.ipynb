## Gork Model Integration

```python
import os
from dotenv import load_dotenv

load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
print(os.getenv('GROQ_MODEL'))
```

```python
from langchain.chat_models import init_chat_model
import os

model = init_chat_model(
    f"groq:{os.getenv('GROQ_MODEL')}"
)

res = model.invoke("Hello, how are you?")
print(res.content)
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.0,
    max_retries=2,
    # other params...
)

res = model.invoke("Hello, how are you?")
print(res.content)
```

## Streamming and batcing

```python
res = model.invoke("Write me a paragraph about Python programming.")
print(res)
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text,end="",flush=True)
```

```python
responses = model.batch(["Why do parrots have colorful feathers?", "Why are some birds able to mimic human speech?", "What is the capital of France?",])
print(responses[0])
for response in responses:
    print(response)
```

```python
responses = model.batch(["Why do parrots have colorful feathers?", "Why are some birds able to mimic human speech?", "What is the capital of France?",],config={"max_concurrency": 5})
print(responses[0])
for response in responses:
    print(response)
```
