# Retrieval-Augmented Generation (RAG)

## What is RAG?

**Retrieval-Augmented Generation (RAG)** is an AI architecture that combines a **Large Language Model (LLM)** with an external knowledge source. Instead of relying only on the model's training data, RAG retrieves relevant information from documents or databases and uses it as context to generate accurate and up-to-date responses.

---

## Why Use RAG?

Traditional LLMs have several limitations:

- Limited to their training knowledge.
- Cannot access private or enterprise documents by default.
- May generate hallucinations (incorrect information).
- Updating knowledge requires retraining.

RAG overcomes these limitations by retrieving relevant information at runtime.

---

# RAG Architecture

```text
                  User Question
                        │
                        ▼
               Embedding Model
                        │
                        ▼
               Similarity Search
                        │
                        ▼
              Vector Database
                        │
                        ▼
          Relevant Document Chunks
                        │
                        ▼
      Prompt = Context + Question
                        │
                        ▼
             Large Language Model
                        │
                        ▼
                 Final Answer
```

---

# How RAG Works

## Step 1: Document Ingestion

Collect documents from one or more sources.

Examples:

- PDF
- Markdown
- Word Documents
- HTML
- Database Records
- REST APIs
- Internal Wikis
- Company Documentation

The documents become the knowledge base for the AI.

---

## Step 2: Document Chunking

Large documents are split into smaller chunks because LLMs and vector databases work better with smaller pieces of information.

Example:

```text
Employee Handbook.pdf

↓

Chunk 1
Chunk 2
Chunk 3
Chunk 4
```

Typical configuration:

- Chunk Size: **500–1000 characters**
- Chunk Overlap: **50–200 characters**

Chunk overlap preserves context between neighboring chunks.

---

## Step 3: Embedding Generation

Each chunk is converted into a dense numerical vector using an embedding model.

Example:

```text
"LangGraph is an AI agent framework."

↓

[0.24, -0.82, 0.51, ...]
```

Embeddings capture the semantic meaning of text rather than exact keywords.

Popular embedding models:

- OpenAI Embeddings
- Gemini Embeddings
- BAAI BGE
- Nomic Embed
- Ollama Embeddings
- Sentence Transformers

---

## Step 4: Store in a Vector Database

Generated embeddings are stored in a vector database for efficient similarity search.

Popular vector databases:

- FAISS
- Chroma
- PGVector
- Pinecone
- Qdrant
- Milvus
- Weaviate

Each stored vector is linked to its original document chunk.

---

## Step 5: User Query

A user asks a question.

Example:

```text
What is LangGraph?
```

---

## Step 6: Query Embedding

The user's question is converted into an embedding using the same embedding model.

```text
"What is LangGraph?"

↓

[0.19, -0.76, 0.43, ...]
```

---

## Step 7: Similarity Search

The vector database compares the query vector with stored vectors.

Common similarity metrics:

- Cosine Similarity
- Dot Product
- Euclidean Distance

Top matching document chunks are retrieved.

Example:

```text
Top Results

1. LangGraph Overview
2. StateGraph Guide
3. Human-in-the-Loop Documentation
```

---

## Step 8: Prompt Augmentation

Retrieved documents are added to the user's prompt.

Example:

```text
Context:

Chunk 1:
LangGraph is...

Chunk 2:
StateGraph...

Question:
What is LangGraph?
```

This process is called **Prompt Augmentation**.

---

## Step 9: Answer Generation

The LLM reads the retrieved context and generates a grounded answer.

Instead of answering from memory alone, it answers using the provided knowledge.

---

# RAG Workflow

```text
Documents
     │
     ▼
Document Loader
     │
     ▼
Text Splitter
     │
     ▼
Embedding Model
     │
     ▼
Vector Database
     ▲
     │
User Query
     │
     ▼
Embedding Model
     │
     ▼
Similarity Search
     │
     ▼
Retrieved Chunks
     │
     ▼
Prompt Builder
     │
     ▼
LLM
     │
     ▼
Final Response
```

# Main Components

| Component | Purpose |
|-----------|---------|
| Document Loader | Reads documents from different sources |
| Text Splitter | Splits documents into chunks |
| Embedding Model | Converts text into vectors |
| Vector Database | Stores embeddings |
| Retriever | Finds relevant chunks |
| Prompt Builder | Combines retrieved context with the question |
| LLM | Generates the final response |

# Advanced RAG Techniques

## Hybrid Search

Combines:

- Vector Search
- Keyword Search (BM25)

Improves retrieval accuracy.

---

## Multi-Query RAG

The LLM generates multiple search queries.

Example:

```text
Original:
"What is LangGraph?"

Generated Queries:

- Explain LangGraph
- LangGraph Architecture
- What is StateGraph?
```

Results from all queries are merged.

---

## Parent-Child Retrieval

Stores:

- Small chunks for searching
- Larger parent documents for context

This improves both precision and readability.

---

## Contextual Compression

Filters retrieved documents before sending them to the LLM.

Benefits:

- Fewer tokens
- Lower cost
- Better answers

---

## Re-ranking

A reranker model scores retrieved documents again.

Instead of using the first retrieved result, documents are reordered by relevance.

---

## Graph RAG

Uses a knowledge graph instead of only vector search.

Useful when relationships between entities are important.

Example:

```text
Employee

↓

Department

↓

Manager

↓

Projects
```

---

## Agentic RAG

Instead of a fixed retrieval pipeline, an AI agent decides:

- Whether retrieval is needed
- Which knowledge source to use
- Which tools to call
- Whether another search is required

This is commonly implemented with frameworks like **LangGraph**.

# Summary

Retrieval-Augmented Generation (RAG) enhances Large Language Models by retrieving relevant information from external knowledge sources before generating a response. A typical RAG pipeline involves document ingestion, chunking, embedding generation, vector storage, similarity search, prompt augmentation, and answer generation. By combining retrieval with generation, RAG enables AI systems to provide accurate, context-aware, and up-to-date responses using private or domain-specific knowledge without retraining the underlying model.

# Step 1 - Load the PDF

Suppose we have:

```
Employee_Handbook.pdf
```

Code:

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Employee_Handbook.pdf")

documents = loader.load()
```

### What happens?

Before:

```
Employee_Handbook.pdf
```

After:

```
[
 Page 1,
 Page 2,
 Page 3,
 ...
 Page 500
]
```

The PDF becomes readable text.

---

# Step 2 - Split the Document

The handbook has **500 pages**.

The LLM cannot efficiently search such a large document.

We split it.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunks = splitter.split_documents(documents)
```

### Before

```
500 Pages
```

### After

```
Chunk 1

Chunk 2

Chunk 3

Chunk 4

...

Chunk 1200
```

Each chunk is much smaller and easier to search.

---

# Example Chunk

Chunk #254

```
Remote Work Policy

Interns may work remotely
up to 2 days every week
with manager approval.

Remote work requests
must be submitted
through the HR Portal.
```

---

# Step 3 - Convert Text into Embeddings

Computers cannot compare paragraphs directly.

They compare vectors.

```python
from langchain_openai import OpenAIEmbeddings

embedding = OpenAIEmbeddings(
    model="text-embedding-3-small"
)
```

Suppose this paragraph becomes:

```
Remote Work Policy

↓

[0.21, -0.33, 0.74, ...]
```

Another sentence:

```
Work from home for interns

↓

[0.20, -0.35, 0.72, ...]
```

Different words...

Same meaning...

Very similar vectors.

---

# Step 4 - Store Inside pgvector

```python
from langchain_postgres import PGVector

db = PGVector(
    embeddings=embedding,
    collection_name="hr_docs",
    connection="postgresql+psycopg://user:password@localhost:5432/company"
)

db.add_documents(chunks)
```

Now every chunk is stored as an embedding.

```
Chunk

↓

Embedding

↓

PostgreSQL + pgvector
```

---

# What Does the Database Look Like?

| Chunk | Vector |
|--------|--------|
| Leave Policy | [0.23,0.55...] |
| Travel Policy | [0.91,0.11...] |
| Remote Work | [0.45,0.67...] |
| Salary Policy | [0.72,0.18...] |

---

# Step 5 - Employee Asks a Question

```
Can interns work remotely?
```

We convert the question into an embedding.

```
Question

↓

Embedding

↓

Search Similar Vectors
```

---

# Step 6 - Similarity Search

The database compares the question with every chunk.

```
Question

↓

Remote Work Chunk

Similarity = 0.97

Leave Policy

Similarity = 0.42

Travel Policy

Similarity = 0.18
```

The Remote Work chunk is the best match.

---

# Retrieve the Best Chunk

```python
retriever = db.as_retriever(
    search_kwargs={"k":1}
)

docs = retriever.invoke(
    "Can interns work remotely?"
)
```

The retrieved chunk:

```
Interns may work remotely
up to 2 days per week
with manager approval.
```

---

# Step 7 - Build the Prompt

Instead of asking only:

```
Can interns work remotely?
```

We build a richer prompt.

```python
context = docs[0].page_content
```

Prompt:

```
You are an HR assistant.

Answer only using the context below.

Context:

Interns may work remotely
up to 2 days every week
with manager approval.

Question:

Can interns work remotely?
```

---

# Step 8 - Ask the LLM

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI(model="gpt-4.1")

prompt = ChatPromptTemplate.from_template("""
Answer only using the provided context.

Context:
{context}

Question:
{question}
""")

chain = prompt | llm

response = chain.invoke({
    "context": context,
    "question": "Can interns work remotely?"
})

print(response.content)
```

Output:

```
Yes.

According to the Employee Handbook,
interns may work remotely
up to 2 days per week
with manager approval.
```

---

## Load PDF

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("employee_handbook.pdf")

documents = loader.load()
```

---

## Split Documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = splitter.split_documents(documents)
```

---

## Create Vector Database

```python
from langchain_openai import OpenAIEmbeddings
from langchain_postgres import PGVector

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-small"
)

vector_store = PGVector(
    embeddings=embeddings,
    collection_name="company_docs",
    connection="postgresql+psycopg://user:password@localhost:5432/rag_db"
)

vector_store.add_documents(chunks)
```

---

## Retrieve Documents

```python
retriever = vector_store.as_retriever(
    search_kwargs={"k": 3}
)

docs = retriever.invoke(
    "Can interns work remotely?"
)
```

---

## Ask the LLM

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4.1")

context = "\n".join(doc.page_content for doc in docs)

prompt = ChatPromptTemplate.from_template("""
Use only the given context.

Context:
{context}

Question:
{question}
""")

chain = prompt | llm

response = chain.invoke({
    "context": context,
    "question": "Can interns work remotely?"
})

print(response.content)
```

## Key Takeaway

RAG does **not** make the LLM smarter. Instead, it gives the LLM access to the **right information at the right time**. By retrieving relevant documents before generating an answer, RAG produces responses that are more accurate, up-to-date, and grounded in your own knowledge base.

```python
from langchain_openai import OpenAIEmbeddings

embedding = OpenAIEmbeddings(
    model="text-embedding-3-small"
)
```

Suppose this paragraph becomes:

```
Remote Work Policy

↓

[0.21, -0.33, 0.74, ...]
```

Another sentence:

```
Work from home for interns

↓

[0.20, -0.35, 0.72, ...]
```

Different words...

Same meaning...

Very similar vectors.

---

# Step 4 - Store Inside pgvector

```python
from langchain_postgres import PGVector

db = PGVector(
    embeddings=embedding,
    collection_name="hr_docs",
    connection="postgresql+psycopg://user:password@localhost:5432/company"
)

db.add_documents(chunks)
```

Now every chunk is stored as an embedding.

```
Chunk

↓

Embedding

↓

PostgreSQL + pgvector
```

---

# What Does the Database Look Like?

| Chunk | Vector |
|--------|--------|
| Leave Policy | [0.23,0.55...] |
| Travel Policy | [0.91,0.11...] |
| Remote Work | [0.45,0.67...] |
| Salary Policy | [0.72,0.18...] |

---

# Step 5 - Employee Asks a Question

```
Can interns work remotely?
```

We convert the question into an embedding.

```
Question

↓

Embedding

↓

Search Similar Vectors
```

---

# Step 6 - Similarity Search

The database compares the question with every chunk.

```
Question

↓

Remote Work Chunk

Similarity = 0.97

Leave Policy

Similarity = 0.42

Travel Policy

Similarity = 0.18
```

The Remote Work chunk is the best match.

---

# Retrieve the Best Chunk

```python
retriever = db.as_retriever(
    search_kwargs={"k":1}
)

docs = retriever.invoke(
    "Can interns work remotely?"
)
```

The retrieved chunk:

```
Interns may work remotely
up to 2 days per week
with manager approval.
```

---

# Step 7 - Build the Prompt

Instead of asking only:

```
Can interns work remotely?
```

We build a richer prompt.

```python
context = docs[0].page_content
```

Prompt:

```
You are an HR assistant.

Answer only using the context below.

Context:

Interns may work remotely
up to 2 days every week
with manager approval.

Question:

Can interns work remotely?
```

---

# Step 8 - Ask the LLM

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI(model="gpt-4.1")

prompt = ChatPromptTemplate.from_template("""
Answer only using the provided context.

Context:
{context}

Question:
{question}
""")

chain = prompt | llm

response = chain.invoke({
    "context": context,
    "question": "Can interns work remotely?"
})

print(response.content)
```

Output:

```
Yes.

According to the Employee Handbook,
interns may work remotely
up to 2 days per week
with manager approval.
```

# Why Not Send the Entire PDF?

The handbook has **500 pages**.

Sending all of it to the LLM would:

- Cost more tokens 💰
- Be slower 🐢
- Exceed the model's context window 📏
- Include irrelevant information ❌

Instead, RAG retrieves only the **most relevant section**, making responses faster, cheaper, and more accurate.

---

# Real Benefits for the Company

Before RAG:

- HR manually answered 500 questions/day.
- Employees waited hours for responses.
- Answers could be inconsistent.

After RAG:

- Employees get answers instantly (24/7).
- Responses come directly from the latest HR handbook.
- HR focuses on complex cases instead of repetitive questions.
- Updating policies is easy—replace the document and re-index it; no model retraining is needed.

# Key Learning

RAG is **not about teaching the LLM everything**.

It is about:

1. **Finding** the right information.
2. **Giving** that information to the LLM.
3. **Generating** an answer based on the retrieved context.

In simple terms:

```
Question

↓

Search

↓

Retrieve

↓

Answer
```

That's the core idea behind **Retrieval-Augmented Generation (RAG)**.
